It was decided to build an Cyano-specific HMM for CAOs using the known Prochlorococcus CAOs and closely related members identified through BLAST.

CAE19267 was blasted against the IMG_V400 database and the top 100 hits collected. Now we check them for length

```python
! cat ../data/img.CAE19267.top.100 | fastaToTab | awk -F '\t' '{ if(length($2)>419 && length($2) < 450) print $0}' | tabToFasta > ../data/img.CAE19267.top.100.len.filtered
! grep -c '>' ../data/img.CAE19267.top.100.len.filtered
```

```python
import os.path
from ete2 import Tree, TreeStyle
def tree_pipeline(infile):
    filename = os.path.splitext(infile)[0]
    !/local1/tools/muscle -in $infile -out $filename\.al
    !cat $filename\.al | sed 's/sp|//g' | sed 's/tr|//g' | sed 's/gi|//g' | sed 's/[(),$+]//g' | sed 's/[|.:;]/_/g' > $filename\.head.al
    !cat $filename\.head.al |  fasta2phy | phyToFasta.pl > $filename\.cln.al
    seqs=!grep -c ">" $filename\.cln.al
    b1 = int(seqs[0])/2+1
    b2 = int(seqs[0])/2+1
    b3 = int(seqs[0])/2
    !/local1/tools/Gblocks_0.91b/Gblocks $filename\.cln.al -e=.fst -b1=$b1 -b2=$b2 -b3=$b3 -b4=2 -b5=h
    !mv $filename\.cln.al.fst $filename\.cln.gb
    !cat $filename\.cln.gb | fasta2phy > $filename\.cln.gb.phy
```

```python
tree_pipeline('../data/img.CAE19267.top.100.len.filtered')
```

```python
!/local1/tools/FastTreeMP $filename\.cln.gb.phy > $filename\.cln.gb.tree
from ete2 import Tree, TreeStyle
t = Tree('%s.cln.gb.tree' % filename)
t.set_outgroup(t.get_common_ancestor("Q5N0P9_Q5N",  "Q31L75_Q31"))
ts = TreeStyle()
ts.show_branch_support = True
ts.scale = 120
t.render(file_name="%%inline", w=600, tree_style=ts) 
```
