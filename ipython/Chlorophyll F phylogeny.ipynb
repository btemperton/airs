```python
best_models = pd.read_csv('../data/combined.best_models', sep='\t')
best_models
```

Now let's find out the abundance of the models across the sequences to identify good candidates

```python
import operator
grouped = best_models.groupby('name')
sfams = dict()
for name, group in grouped:
    for m in group.best_model:
        try:
            sfams[m] = sfams[m] +1
        except KeyError:
            sfams[m] = 1

sfams = sorted(sfams.iteritems(), key=operator.itemgetter(1))
sfams
```

So, we have 17 SFams for CAOs from this paper. Now we can scan the <em>C. fritschii</em> predicted proteins with these SFams to identify putative members. Putative proteins were downloaded from IMG/ER (taxon id. 2509601037).  

```python
with open('../data/satoh.cao.models', 'w') as handle:
    for k in sfams:
        handle.write('%s\n' % k[0])
!hmmfetch -f /local1/data/sfam/Sfam.hmm ../data/satoh.cao.models > ../data/cao_sfams_for_search
!hmmpress ../data/cao_sfams_for_search
!hmmscan --cpu 6 -o ../data/c.fritschii.caos --tblout ../data/c.fritschii.caos.tbl -E 0.01 ../data/cao_sfams_for_search ../data/11048.faa
```

```python
def hmm_tbl_to_dataframe(filename):
    df = pd.read_csv(filename, sep='\s+', skiprows=3, header=None)
    df = df.drop([1,3,10,11,12,13,14,15,16,17,18], 1)
    df.columns = ['target_name','query_name', 'full_e_val', 'full_score', 'full_bias', 'best_e_val', 'best_score', 'best_bias', ]
    return df

cao_df = hmm_tbl_to_dataframe('../data/c.fritschii.caos.tbl')
cao_groups = cao_df.groupby('target_name')
```

```python
import os.path
from ete2 import Tree, TreeStyle
def tree_pipeline(infile):
    filename = os.path.splitext(infile)[0]
    !/local1/tools/muscle -in $infile -out $filename\.al
    !cat $filename\.al | sed 's/sp|//g' | sed 's/tr|//g' | sed 's/gi|//g' | sed 's/[(),$+]//g' | sed 's/[|.:;]/_/g' > $filename\.head.al
    !cat $filename\.head.al |  fasta2phy | phyToFasta.pl > $filename\.cln.al
    seqs=!grep -c ">" $filename\.cln.al
    b1 = int(seqs[0])/2+1
    b2 = int(seqs[0])/2+1
    b3 = int(seqs[0])/2
    !/local1/tools/Gblocks_0.91b/Gblocks $filename\.cln.al -e=.fst -b1=$b1 -b2=$b2 -b3=$b3 -b4=2 -b5=h
    !mv $filename\.cln.al.fst $filename\.cln.gb
    !cat $filename\.cln.gb | fasta2phy > $filename\.cln.gb.phy
    !/local1/tools/FastTreeMP $filename\.cln.gb.phy > $filename\.cln.gb.tree
```

```python
tree_pipeline('../data/tree1.faa')
```

```python
from ete2 import Tree, TreeStyle
t = Tree('../data/tree1.cln.gb.tree')
ts = TreeStyle()
ts.show_branch_support = True
ts.scale = 120
t.render(file_name="%%inline", w=600, tree_style=ts)
```
